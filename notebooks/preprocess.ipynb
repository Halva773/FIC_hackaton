```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
data = pd.read_csv('../data/train_data.csv', index_col=0)
print(data.shape)
data.head()
```

```python
data = data.drop(columns=['position', 'key_skills', 'salary', 'unique_work', 'work_experience'])
categorial_features = ['country', 'city', 'client_name', 'company_type']
for feature in categorial_features:
    data[feature] = data[feature].astype('category').cat.codes
data['grade_proof'] = data['grade_proof'].apply(lambda x: 1 if x == 'подтверждён' else 0)
```

```python
import pandas as pd
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
```

```python
target_column = 'grade_proof'

# Разделение на X (признаки) и y (целевая переменная)
target_column = 'grade_proof'
X = data.drop(columns=[target_column])
y = data[target_column]

def normalize_column(col):
    # Если стандартное отклонение колонки 0, не нормализуем её
    if col.std() == 0:
        return col
    return (col - col.min()) / (col.max() - col.min())

# Применение нормализации к каждому столбцу
X = X.apply(normalize_column)

X.shape, y.shape
```

```python
# X = X.fillna(X.mode().iloc[0])

# Разделение на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42, stratify=y)

# Применение SMOTE для балансировки классов
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

# Вывод размеров выборок
X_resampled.shape, X_test.shape
```

```python
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score

# Создание модели CatBoostClassifier
model = CatBoostClassifier(
    iterations=1000,       # Количество итераций
    depth=6,               # Глубина деревьев
    learning_rate=0.1,     # Скорость обучения
    loss_function='Logloss', # Логистическая регрессия (классификация)
    verbose=200            # Выводить информацию о ходе обучения каждые 200 итераций
)

# Обучение модели
model.fit(X_resampled, y_resampled)

# Прогнозирование на тестовых данных
y_pred = model.predict(X_test)

# Оценка модели
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.4f}')
y_pred_loaded = [i[1] for i in model_loaded.predict_proba(X_test)]
print(f'ROC_AUC (CatBoost): {roc_auc_score(y_test, y_pred_loaded)}')

# Дополнительные метрики
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Матрица ошибок
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
import optuna
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
```

```python
def objective(trial):
    params = {
        'depth': trial.suggest_int('depth', 4, 10),
        'learning_rate': trial.suggest_loguniform('learning_rate', 1e-3, 1e-1),
        'l2_leaf_reg': trial.suggest_loguniform('l2_leaf_reg', 1e-8, 10.0),
        'random_strength': trial.suggest_uniform('random_strength', 1e-9, 10),
        'bagging_temperature': trial.suggest_uniform('bagging_temperature', 0.0, 1.0),
        'od_type': trial.suggest_categorical('od_type', ['IncToDec', 'Iter']),
        'border_count': trial.suggest_int('border_count', 32, 255),
        'grow_policy': trial.suggest_categorical('grow_policy', ['SymmetricTree', 'Depthwise']),
    }

    model = CatBoostClassifier(**params, verbose=False)
    model.fit(
        X_train,
        y_train,
        eval_set=(X_test, y_test),
        early_stopping_rounds=100,
        use_best_model=True
    )

    preds = model.predict(X_test)
    score = accuracy_score(y_test, preds)

    return score

# Запуск процесса оптимизации
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=50)

# Получение лучших параметров
best_params = study.best_params
print(f'Лучшие параметры: {best_params}')

# Обучение финальной модели с лучшими параметрами
model = CatBoostClassifier(**best_params, verbose=False)
model.fit(
    X_train,
    y_train,
    eval_set=(X_test, y_test),
    early_stopping_rounds=100,
    use_best_model=True
)

preds = model.predict(X_test)
score = accuracy_score(y_test, preds)
print(f'Точность на валидационном наборе: {score:.4f}')
```

```python
# Прогнозирование на тестовых данных
y_pred = model.predict(X_test)

# Оценка модели
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.4f}')
y_pred_loaded = [i[1] for i in model.predict_proba(X_test)]
print(f'ROC_AUC (CatBoost): {roc_auc_score(y_test, y_pred_loaded)}')

# Дополнительные метрики
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Матрица ошибок
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```
